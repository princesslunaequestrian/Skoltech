Final answer:
$$
F S(\sin x)=\frac{2}{\pi}-\frac{4}{\pi} \sum_{n=1}^{\infty} \frac{1}{4 n^{2}-1} \cos 2 k x
$$

$$
F S(\cos x)=\frac{8}{\pi} \sum_{n=1}^{\infty} \frac{n}{4 n^{2}-1} \sin 2 n x
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi, 1000)


def FS_sin(x, m):
    sum = np.zeros(x.shape)
    for i in range(1, m + 1):
        sum = sum + (2 * np.cos(2 * i * x)) / (4 * (i)**2 - 1)
    return (2 / np.pi) * (1 - sum)


def FS_cos(x, m):
    sum = np.zeros(x.shape)
    for i in range(1, m + 1):
        sum = sum + (i * np.sin(2 * i * x)) / (4 * (i)**2 - 1)
    return (8 / np.pi) * (sum)


plt.rcParams.update(
    {'font.size': 12, 'lines.markersize': 0, 'lines.linewidth': 2, 'lines.marker': '.', 'axes.grid': True})
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x, np.sin(x), label='$f(x) = sin(x)$')
ax1.plot(x, FS_sin(x, 2), label='FS (2-term)')
ax1.plot(x, FS_sin(x, 10), label='FS (10-term)')
ax1.set_xlim(0, 3.15)
ax1.legend(fontsize=14)

ax2.plot(x, np.cos(x), label='$f(x) = cos(x)$')
ax2.plot(x, FS_cos(x, 2), label='FS (2-term)')
ax2.plot(x, FS_cos(x, 10), label='FS (10-term)')
ax2.set_xlim(0, 3.15)
ax2.legend(fontsize=14)
```

$sin(x)$ is approximated with Fourier series significantly better than $cos(x)$ is

The reason is: Fourier series approximation for the function begins with $FS(0) = 0$ and the value of $cos(0) = 1$

Final answer: 
$\begin{aligned}
X_{G}(t)=&-\frac{1}{\pi i} \sum\frac{1}{k} \frac{1-(-1)^{k}}{\omega_{0}^{2}-k^{2}} \cdot \cos \omega_{0} t-\frac{1}{\pi \omega_{0}} \sum \frac{1-(-1)^{k}}{\omega_{0}^{2}-k^{2}} \cdot \sin \omega_{0} t \\
&+\frac{1}{\pi i} \sum \frac{1}{k} \frac{1-(-1)^{k}}{\omega_{0}^{2}-k^{2}} e^{ik t}
\end{aligned}$

```python
import numpy as np

w0 = np.sqrt(2)
T = np.arange(0, 10*np.pi, 0.1)

K = np.arange(1,9)
# print(K,'\n',T)
```

```python
def C1(k):
    return -1/(np.pi*1j) * np.sum(1 / k * (1-(-1)**k) / (w0**2 - k**2))
    

def C2(k):
    return -1/(np.pi*w0) * np.sum((1-(-1)**k) / (w0**2 - k**2))


def g_kt(k,t):
    return 1 / k * (1-(-1)**k) / (w0**2 - k**2) * np.exp(1j*k*t)


X_gen = np.zeros(T.shape, dtype='complex_')

for i in range(len(T)):
    X_gen[i] = C1(K) * np.cos(w0*T[i]) + C2(K) * np.sin(w0*T[i])\
         + 1/(np.pi*1j) * np.sum(g_kt(K, T[i]))

```

This what happens when $\omega_0$ is very close to $1$. \
From physical point of view we observe the motion of simple pendulum in the case of small displacements
$$ \ddot x + \omega_0 x = force$$

In our case the external force exhibits very strange behaviour. This behaviour can be described by this Fourier series:
$$
f(t)=\sum_{k=-\infty}^{k=\infty} \frac{1}{\pi i k}\left(1-(-1)^{k}\right) e^{i k t}=(-1)^{m}
$$

The movement, in tern, is described by:
$\begin{aligned}
X_{G}(t)=&-\frac{1}{\pi i} \sum_{k=-\infty}^{k=\infty} \frac{1}{k} \frac{1-(-1)^{k}}{\omega_{0}^{2}-k^{2}} \cdot \cos \omega_{0} t-
\frac{1}{\pi \omega_{0}} \sum_{k=-\infty}^{k=\infty} \frac{1-(-1)^{k}}{\omega_{0}^{2}-k^{2}} \cdot \sin \omega_{0} t \\
&+\frac{1}{\pi i} \sum_{k=-\infty}^{k=\infty} \frac{1}{k} \frac{1-(-1)^{k}}{\omega_{0}^{2}-k^{2}} e^{ik t}
\end{aligned}$

If $\omega_0$ will equal $1$ one can observe the resonanse case: when the frequency of oscillations is equal to the frequency of the external force.

From point of mathematics I suppose that Sums present in $X_G(t)$ will no longer converge. As a result the energy of every next term in the series will be bigger than the previous one. Thus the power spectrum will grow infinitely.

Define Functions for Jacobian, gradient, partial derrivatives:


```python
#fit function
def func(x, a, b):
    return a*(1 - np.exp(-b*x))


def dfda(x, a, b):
    return 1 - np.exp(-b*x)


def dfdb(x, a, b):
    return a*x*np.exp(-b*x)

#rms function


def J(x, a, b, y):
    n = len(x)
    f = func(x, a, b)
    return np.sqrt(1/n)*np.sqrt(np.sum(np.power(f - y, 2)))
```

```python
def faa(x, a, b):

    return 0

def fbb(x, a, b):

    return -a*np.power(x, 2)*np.exp(-b*x)

def fab(x, a, b):

    return x*np.exp(-b*x)

def J(x, a, b, y):
    n = len(x)
    f = func(x, a, b)

    return 1/(2*n)*np.sum(np.power(f - y, 2))

def dJda(x, a, b, y):
    n = len(x)
    f = func(x, a, b)

    return 1/n*np.sum((f - y)*dfda(x, a, b))

def dJdb(x, a, b, y):
    n = len(x)
    f = func(x, a, b)

    return 1/n*np.sum((f - y)*dfdb(x, a, b))

def Jaa(x, a, b, y):
    n = len(x)
    f = func(x, a, b)
    fa = dfda(x, a, b)

    return 1/n * np.sum(fa*fa + (f - y)*faa(x, a, b))

def Jbb(x, a, b, y):
    n = len(x)
    f = func(x, a, b)
    fb = dfdb(x, a, b)

    return 1/n * np.sum(fb*fb + (f - y)*fbb(x, a, b))

def Jab(x, a, b, y):
    n = len(x)
    f = func(x, a, b)
    fa = dfda(x, a, b)
    fb = dfdb(x, a, b)
    f2 = fab(x, a, b)

    return 1/n * np.sum(fa*fb + f2*(f - y))

def grad(x, a, b):
    G = np.zeros(2)

    G[0] = dJda(x, a, b, y)
    G[1] = dJdb(x, a, b, y)

    return G



def HJ(x, a, b, y):
    H = np.zeros((2, 2))
    H[0][0] = Jaa(x, a, b, y)
    H[1][0] = H[0][1] = Jab(x, a, b, y)
    H[1][1] = Jbb(x, a, b, y)

    return H
```

MSE Cost function:

```python
a = np.arange(0, 4, 0.01)
b = np.arange(0, 4, 0.01)

A, B = np.meshgrid(a, b)

HH = np.zeros(A.shape)

for i in range(HH.shape[0]):
    for j in range(HH.shape[1]):
        HH[i][j] = J(x, A[i][j], B[i][j], y)
```

```python
iters = 50000
target_accuracy = 0.005

points = np.zeros((iters, 2))

p0 = np.array([1, 1])

for i in range(iters):

    points[i, :] = p0

    inv = np.linalg.inv(HJ(x, p0[0], p0[1], y))

    scale = i if i else 1

    p = p0 - inv.dot(grad(x, p0[0], p0[1]))/np.sqrt(scale)/100

    if (J(x, p[0], p[1], y) > J(x, p[0], p[1], y)):
        print('Divergence started to occur at step {}'.format(i))
        print('J = {:0,.3f} at a={:1,.2f}, b={:2,.2f}'.format(J(x, p0[0], p0[1], y), p0[0], p0[1]))
        break

    if (J(x, p[0], p[1], y) < target_accuracy):
        print('Reasonable accuracy reached')
        print('J = {:0,.3f} at a={:1,.2f}, b={:2,.2f}'.format(J(x, p[0], p[1], y), p[0], p[1]))
        break

    p0 = p

if (i == iters-1):
    print('Iterations expired')
    print('J = {:0,.3f} at a={:1,.2f}, b={:2,.2f}'.format(J(x, p[0], p[1], y), p[0], p[1]))

    
```
