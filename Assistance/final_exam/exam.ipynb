```python
# YOUR CODE HERE
with open('./inputs/2/circuits.csv') as f:
    Circ = pd.read_csv(f)

with open('./inputs/2/drivers.csv', encoding="utf-8") as f:
    Driv = pd.read_csv(f)

with open('./inputs/2/races.csv') as f:
    Rac = pd.read_csv(f)

with open('./inputs/2/results.csv') as f:
    Res = pd.read_csv(f)
```

```python
ResRac = Res.merge(Rac, on='raceId', how='inner')
ResRac = ResRac[ResRac['year'] >= 2021].sort_values(by='driverId')
ResRac
```

```python
Ranking = pd.DataFrame(columns=['driverId', 'points'])

driverIds = pd.unique(ResRac['driverId'])
points = {}

for id in driverIds:
    points[id] = ResRac[ResRac['driverId'] == id]['points'].sum()

Ranking = pd.DataFrame.from_dict(points, orient='index', columns=['points'])
Ranking['driverId'] = Ranking.index
Ranking = Ranking.merge(Driv, on='driverId', how='inner')
Ranking = Ranking.loc[:, ['points', 'driverId', 'surname']]
Ranking = Ranking.sort_values(by='points', ascending=False)
Ranking
```

```python
Ranking = Ranking['surname'].reset_index(drop=True)
Ranking.index = Ranking.index + 1
Ranking.to_csv('./2.out', sep='\t', header=False)
```
