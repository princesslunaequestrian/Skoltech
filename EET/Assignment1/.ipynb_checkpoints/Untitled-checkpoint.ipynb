```python
import matplotlib.pyplot as plt
from datetime import timedelta, datetime
```

```python

%matplotlib qt
 
fig, ax = plt.subplots()
 
ax.set_xlim([0,7])
ax.set_ylim([-1,1])
 
A_pad_x = 0
A_pad_y = 0
B_pad_x = 6.8
B_pad_y = 0
 
puck_x = 3.5
puck_y = 0
puck_angle = 0
 
puck_vx = 0.05
puck_vy = 0.03

rotationw = 0

#prevtime = datetime.now()
 
A_pad_disp, = ax.barh(A_pad_y, 0.2, height = 0.3, animated = True, color = 'k', edgecolor = 'b')
B_pad_disp, = ax.barh(B_pad_y, 0.2, height = 0.3, animated = True, color = 'k', edgecolor = 'b')
puck_disp = ax.scatter([puck_x], [puck_y], s = 200, c = 'g', marker = 'D', animated = True)
 
A_pad_disp.set_x(A_pad_x)
B_pad_disp.set_x(B_pad_x)
 
def contains(disp, x, y):
    return disp.get_bbox().contains(x,y)
 
def puck_update(puck_x, puck_y, puck_vy, puck_vx, puck_angle, rotationw):
    
    puck_x = puck_x + puck_vx
    puck_y += puck_vy
    puck_angle += rotationw
    
    if puck_y > 1 or puck_y < -1:
        puck_vy = puck_vy*-1
    
    if puck_x > 7 or puck_x < 0:
        puck_x = 3.5
        puck_y = 0
        puck_vx = 0.05
        puck_vy = 0.03
        
    if contains(A_pad_disp, puck_x, puck_y) or contains(B_pad_disp, puck_x, puck_y):
        
        if (puck_vx*puck_vy < 0):
            rotationw -= 1
        elif (puck_vx*puck_vy > 0):
            rotationw += 1
        elif (puck_vx*puck_vy == 0):
            rotationw = 0
        
        puck_vx *= -1        
    
    return puck_x, puck_y, puck_vy, puck_vx, puck_angle, rotationw

def draw(background):
    global puck_x, puck_y, puck_vy, puck_vx, puck_angle, rotationw
 
    ax.figure.canvas.restore_region(background)
    
    puck_x, puck_y, puck_vy, puck_vx, puck_angle, rotationw = puck_update(puck_x, puck_y, puck_vy, puck_vx, puck_angle, rotationw)
    puck_disp.set_offsets([[puck_x, puck_y]])
    
    A_pad_disp.set_x(A_pad_x)
    B_pad_disp.set_x(B_pad_x)
    A_pad_disp.set_y(A_pad_y)
    B_pad_disp.set_y(B_pad_y)
    
    angle_disp = ax.text(5.6, 0.86, s=str(puck_angle), animated=True)
    
    ax.draw_artist(A_pad_disp)
    ax.draw_artist(B_pad_disp)
    ax.draw_artist(puck_disp)
    ax.draw_artist(angle_disp)
    
 
    ax.figure.canvas.blit(ax.bbox)
    ax.figure.canvas.flush_events()

def on_key_press(event):
    
    global A_pad_y, B_pad_y, puck_vy, prevtime
    
    if event.key == 'e':
        A_pad_y = A_pad_y + 0.03
        if (A_pad_y > 0.7):
            A_pad_y = 0.7
        if contains(A_pad_disp, puck_x, puck_y) or contains(B_pad_disp, puck_x, puck_y):
            puck_vy += 0.03
        
    if event.key == 'd':
        A_pad_y = A_pad_y - 0.03
        if (A_pad_y < -1):
            A_pad_y = -1
        if contains(A_pad_disp, puck_x, puck_y) or contains(B_pad_disp, puck_x, puck_y):
            puck_vy -= 0.03
        
    if event.key == 'i':
        B_pad_y = B_pad_y + 0.03
        if (B_pad_y > 0.7):
            B_pad_y = 0.7
        if contains(A_pad_disp, puck_x, puck_y) or contains(B_pad_disp, puck_x, puck_y):
            puck_vy += 0.03
        
    if event.key == 'k':
        B_pad_y = B_pad_y - 0.03
        if (B_pad_y < -1):
            B_pad_y = -1
        if contains(A_pad_disp, puck_x, puck_y) or contains(B_pad_disp, puck_x, puck_y):
            puck_vy -= 0.03
        
ax.figure.canvas.mpl_connect('key_press_event', on_key_press)
ax.figure.canvas.mpl_disconnect(fig.canvas.manager.key_press_handler_id)
 
#for faster moving 
ax.figure.canvas.draw()
background = ax.figure.canvas.copy_from_bbox(ax.bbox)
 
plt.show()
 
while True:
    draw(background)
```
