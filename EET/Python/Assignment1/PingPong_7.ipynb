```python
%matplotlib qt
 
fig, ax = plt.subplots()
 
ax.set_xlim([0,7])
ax.set_ylim([-1.5,1.5])

top_col = False
bot_col = False
A_touch = False
B_touch = False
 
A_pad_x = 0
A_pad_y = 0
B_pad_x = 6.8
B_pad_y = 0

A_pad_vy = 0
B_pad_vy = 0
 
puck_x = 3.5
puck_y = 0
puck_angle = 0

puck_R = 1 #puck Radius
 
puck_vx = 0.05
puck_vy = 0.03
puck_w = 0

ex = 0.75
ey = 1
alpha = 2/5

A_pad_disp, = ax.barh(A_pad_y, 0.2, height = 0.3, animated = True, color = 'k', edgecolor = 'b')
B_pad_disp, = ax.barh(B_pad_y, 0.2, height = 0.3, animated = True, color = 'k', edgecolor = 'b')
puck_disp = ax.scatter([puck_x], [puck_y], s = 200, c = 'g', marker = 'D', animated = True)
 
A_pad_disp.set_x(A_pad_x)
B_pad_disp.set_x(B_pad_x)
 
def contains(disp, x, y):
    return disp.get_bbox().contains(x,y)
 
def puck_update(puck_x, puck_y, puck_angle, puck_vx, puck_vy, puck_w, A_pad_y, B_pad_y, A_pad_vy ,B_pad_vy):
    
    global top_col, bot_col, leave, A_touch, B_touch
    
    puck_x += puck_vx
    puck_y += puck_vy
    puck_angle += puck_w
    
    A_pad_y += A_pad_vy
    B_pad_y += B_pad_vy
    
    if (puck_y >= 1.5):
        #top collision
        top_col = True
        
    elif (puck_y <= -1.5):
        #bot collision
        bot_col = True
        
    elif (puck_x >= 7 ) or (puck_x <= 0):
        #leave borders
        puck_x = 3.5
        puck_y = 0
        puck_vx = 0.1
        puck_vy = 0.05
        
    elif contains(A_pad_disp, puck_x, puck_y):
        #touch A pad
        A_touch = True
        
    elif contains(B_pad_disp, puck_x, puck_y):
        #touch B pad
        B_touch = True
        
    if (A_pad_y >= 1.2):
        A_pad_y = 1.2
        A_pad_vy = 0
    if (A_pad_y <= -1.5):
        A_pad_y = -1.5
        A_pad_vy = 0
    if (B_pad_y >= 1.2):
        B_pad_y = 1.2
        B_pad_vy = 0
    if (B_pad_y <= -1.5):
        B_pad_y = -1.5
        B_pad_vy = 0
    
    return puck_x, puck_y, puck_angle, A_pad_y, B_pad_y, A_pad_vy, B_pad_vy

def collision(puck_vx, puck_vy, puck_w, A_pad_vy, B_pad_vy):
    
    global top_col, bot_col, A_touch, B_touch, alpha, ex, ey
    
    puck_vx2 = puck_vx
    puck_vy2 = puck_vy
    puck_w2 = puck_w
    puck_tang = puck_R * puck_w
    
    if (top_col):
        #top collision
        puck_w2 = (-puck_w*(alpha-ex) + puck_vx + puck_vx*ex)/((1+alpha)*puck_R)   
        puck_vx2 = puck_vx*((1 - alpha*ex)/(1 + alpha) + alpha*(1 + ex)/(1 + alpha)*(puck_tang/puck_vx))
        puck_vy2 = -ey*puck_vy
        
        print('Top touch: \n vx2 = {} \n w2 = {} \n vy2 = {} \n'.format(puck_vx2, puck_w2, puck_vy2))
        top_col = False
        
    if (bot_col):
        puck_w2 = (puck_w*(alpha-ex) + puck_vx + puck_vx*ex)/((1+alpha)*puck_R)   
        puck_vx2 = puck_vx*((1 - alpha*ex)/(1 + alpha) + alpha*(1 + ex)/(1 + alpha)*(puck_tang/puck_vx))
        puck_vy2 = -ey*puck_vy
        
        print('Bot touch: \n vx2 = {} \n w2 = {} \n vy2 = {} \n'.format(puck_vx2, puck_w2, puck_vy2))
        bot_col = False
        
    if (A_touch):
        relative_vy = puck_vy - A_pad_vy
        if (relative_vy == 0):
            puck_vx2 = -puck_vx
            puck_vy2 = puck_vy
        else:
            puck_w2 = (puck_w*(alpha-ex) + relative_vy + relative_vy*ex)/((1+alpha)*puck_R)   
            puck_vy2 = relative_vy*((1 - alpha*ex)/(1 + alpha) + alpha*(1 + ex)/(1 + alpha)*(puck_tang/relative_vy))
            puck_vx2 = -ey*puck_vx

        print('A touch: \n vx2 = {} \n w2 = {} \n vy2 = {} \n'.format(puck_vx2, puck_w2, puck_vy2))
        A_touch = False
        
        
    if (B_touch):
        relative_vy = puck_vy - B_pad_vy
        if (relative_vy == 0):
            puck_vx2 = -puck_vx
            puck_vy2 = puck_vy
        
        else:
            puck_w2 = (-puck_w*(alpha-ex) + relative_vy + relative_vy*ex)/((1+alpha)*puck_R)   
            puck_vy2 = relative_vy*((1 - alpha*ex)/(1 + alpha) + alpha*(1 + ex)/(1 + alpha)*(-puck_tang/relative_vy))
            puck_vx2 = -ey*puck_vx
        
        print('B touch: \n vx2 = {} \n w2 = {} \n vy2 = {} \n'.format(puck_vx2, puck_w2, puck_vy2))
        B_touch = False
        
    return puck_vx2, puck_vy2, puck_w2
    
    

def draw(background):
    global puck_x, puck_y, puck_vy, puck_vx, puck_angle, puck_w, A_pad_y, B_pad_y, A_pad_vy, B_pad_vy
 
    ax.figure.canvas.restore_region(background)
    
    puck_x, puck_y, puck_angle, A_pad_y, B_pad_y, A_pad_vy, B_pad_vy = puck_update(puck_x, puck_y, puck_angle,
                                                               puck_vx, puck_vy, puck_w,
                                                              A_pad_y, B_pad_y, A_pad_vy, B_pad_vy)
    puck_vx, puck_vy, puck_w = collision(puck_vx, puck_vy, puck_w, A_pad_vy, B_pad_vy)
    
    puck_disp.set_offsets([[puck_x, puck_y]])
    A_pad_disp.set_x(A_pad_x)
    B_pad_disp.set_x(B_pad_x)
    A_pad_disp.set_y(A_pad_y)
    B_pad_disp.set_y(B_pad_y)
    
    angle_disp = ax.text(3, 0.86, s=str(puck_angle)+':'+str(puck_w), animated=True)
    speed_disp = ax.text(3, 0.66, s=str(puck_vx)+':'+str(puck_vy), animated=True)
    
    ax.draw_artist(A_pad_disp)
    ax.draw_artist(B_pad_disp)
    ax.draw_artist(puck_disp)
    ax.draw_artist(angle_disp)
    ax.draw_artist(speed_disp)
    
 
    ax.figure.canvas.blit(ax.bbox)
    ax.figure.canvas.flush_events()

def on_key_press(event):
    
    global A_pad_y, B_pad_y, A_pad_vy, B_pad_vy
    
    if event.key == 'e':
        A_pad_vy += 0.03
        
    if event.key == 'd':
        A_pad_vy -= 0.03
       
    if event.key == 'i':
        B_pad_vy += 0.03       
        
    if event.key == 'k':
        B_pad_vy -= 0.03

        
ax.figure.canvas.mpl_connect('key_press_event', on_key_press)
ax.figure.canvas.mpl_disconnect(fig.canvas.manager.key_press_handler_id)
 
#for faster moving 
ax.figure.canvas.draw()
background = ax.figure.canvas.copy_from_bbox(ax.bbox)
 
plt.show()
 
while True:
    draw(background)
```
